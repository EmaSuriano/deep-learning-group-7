```python
import matplotlib.pyplot as plt
import numpy as np

import pandas as pd
import json

import cv2
import requests
import os


DS_PATH = "ds"
DS_ENTRY = f"{DS_PATH}/Poses.json"
```

## Load DS


```python
with open(DS_ENTRY, "r") as file:
    data = json.load(file)

df = pd.DataFrame(data["Poses"])

display(df.head())
```

```python
# Obtener pose random del DS
idx = np.random.randint(0, len(df))

item = df.iloc[idx]

pose = item["sanskrit_name"]

# Obtengo la carpeta de imagenes de la pose
folder = f"{DS_PATH}/{pose}"

files = os.listdir(folder)

# Obtener imagen random de la pose
file_idx = np.random.randint(0, len(files))

file = f"{folder}/{files[file_idx]}"

img = cv2.imread(file, cv2.COLOR_BGR2RGB)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.title(pose)
plt.imshow(img)

plt.show()
```
